```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from PIL import Image
```

```python
frame = cv2.imread("5E5A1086.JPG")
```

```python
plt.imshow(frame)
```

```python
rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
```

```python
classifier = cv2.CascadeClassifier("haarcascade_frontFace.xml")
```

```python
faces= classifier.detectMultiScale(frame)
```

```python
face = faces[0]
x, y, w, h = face
```

```python
face = rgb[y:y+h , x:x+w]
```

```python
plt.imshow(face)
```

```python
plt.imshow(cv2.rectangle(rgb,(x,y), (x+w, y+h), (0,0,255), 100))
```
